```python
import pymysql #Allows the connection to mysql
import pandas as pd #We need pandas to obtain a Data Frame
import getpass # function to prompt a cell to write down our password
from sqlalchemy import create_engine #Method to produce an engine based on a URL, that includes, host, user....
```

```python
user = getpass.getuser()
password = getpass.getpass(prompt="Please enter your MySQL password",stream=None)
engine = create_engine('mysql+pymysql://{}:{}@localhost:3306/classicmodels'.format(user,password))
connection = pymysql.connect(host='localhost',user=user,password=password, db='classicmodels')                           
```

```python
a="""SELECT * FROM employees;""" #Data frame from SQL
```

```python
employees=pd.read_sql_query(a,engine)
```

```python
df_employees =pd.DataFrame(employees)
df_employees.columns
```

```python
df_employees['officeCode']=df_employees['officeCode'].astype(int)#Converting object into int to Vlookup
```

```python
#Excel data frame, remeber to change directory to yours
df_offices=pd.read_excel('C:\\Users\\jose\\OneDrive - Dublin Business School (DBS)\\Desktop\\Files\\offices_2022-08-07.xlsx')
```

```python
df_offices['officeCode']=df_offices['officeCode'].astype(int) #Converting object into int to Vlookup
```

```python
df_offices.columns
```

```python
df_offices['officeCode']=df_offices['officeCode'].astype(int) #Converting object into int to Vlookup
```

```python
#Vlookuping both data frames using officeCode as a unique identifier
#Here we are adding city and phone into df_employees
df_employees = pd.merge(df_employees,df_offices[['officeCode','city']],on='officeCode',how='left')
df_employees = pd.merge(df_employees,df_offices[['officeCode','phone']],on='officeCode',how='left')
```
