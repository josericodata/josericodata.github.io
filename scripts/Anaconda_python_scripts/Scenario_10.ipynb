```python
import matplotlib.pyplot as plt #Importing this library for plot generation
import geopandas as gpd #To generate maps

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

ax2 = world[world.continent == 'Asia'].plot(figsize=(8,8), edgecolor=u'black', cmap='Pastel1')

world[world.name == "Japan"].plot(edgecolor=u'black', color='Green', ax=ax2)


plt.ylabel('Latitude')
plt.xlabel('Longitude')

plt.show()
```
